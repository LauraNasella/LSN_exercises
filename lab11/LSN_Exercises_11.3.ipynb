## Esercizio 11.3 - Fit funzione trigonometrica 2D

Ora utilizziamo il modello ottimizzato nel precedente esercizio (2 hidden layers, 15 neuroni per layer, f.ne di attivazione 'relu', ottimizzatore 'Nadam', f.ne di loss 'msle') per fittare una funzione trigonometrica 2D:
$$f(x,y) = \sin(x^2+y^2)$$
nel range $x \in [-3/2,3/2]$ e $y \in [-3/2,3/2]$.

```python
import numpy as np

# generate training inputs
np.random.seed(0)
n_train = 20000
n_valid = int(n_train/10)
x_train = np.random.uniform(-1.5, 1.5, n_train)
y_train = np.random.uniform(-1.5, 1.5, n_train)
x_valid = np.random.uniform(-1.5, 1.5, n_valid)
y_valid = np.random.uniform(-1.5, 1.5, n_valid)
#x_valid.sort()
#y_valid.sort()
f_target = np.sin(x_valid**2 + (y_valid**2)) # ideal (target) function

sigma = 0.15 # noise standard deviation
f_train = np.random.normal(np.sin(x_train**2 + (y_train**2)), sigma) # actual measures from which we want to guess regression parameters
f_valid = np.random.normal(np.sin(x_valid**2 + (y_valid**2)), sigma)

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (15,7))
plt.suptitle('', size=20)

ax = fig.add_subplot(1, 2, 1, projection='3d')

ax.scatter(x_valid[:], y_valid[:], f_valid, c=f_valid)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f')
ax.set_title('Punti di validazione con rumore', size=15)

ax = fig.add_subplot(1, 2, 2, projection='3d')

ax.scatter(x_valid[:], y_valid[:], f_target, c=f_target)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Punti di validazione target', size=15)

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
```

```python
#msle
model_2_msle = tf.keras.Sequential()
model_2_msle.add(Dense(15, input_shape=(2,), activation='relu'))
model_2_msle.add(Dense(15, activation='relu'))
model_2_msle.add(Dense(15, activation='relu'))
model_2_msle.add(Dense(1, activation='elu'))

model_2_msle.compile(optimizer='Nadam', loss='msle', metrics=['mse'])

model_2_msle.summary()
```

```python
xy_train=np.zeros((n_train,2))
xy_train[:,0] = x_train 
xy_train[:,1] = y_train
 
xy_valid=np.zeros((n_valid,2))
xy_valid[:,0] = x_valid
xy_valid[:,1] = y_valid

history = model_2_msle.fit(x=xy_train, y=f_train, 
          batch_size=100, epochs=60,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(xy_valid, f_valid))
```

```python
score_2_msle = model_2_msle.evaluate(xy_valid, f_valid, batch_size=100, verbose=1)
# print performance
print()
print('Tesloss:', score_2_msle[0])
print('Test accuracy:', score_2_msle[1])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
xy_predicted=np.zeros((n_valid,2))
xy_predicted[:,0]=np.random.uniform(-1.5, 1.5, n_valid) 
xy_predicted[:,1]=np.random.uniform(-1.5, 1.5, n_valid)

f_predicted = model_2_msle.predict(xy_predicted)

fig = plt.figure(figsize = (15,7))
ax = fig.gca(projection='3d')
ax.plot_trisurf(xy_valid[:,0], xy_valid[:,1], f_target, linewidth=2, antialiased=True, alpha=0.4)
ax.scatter(xy_predicted[:,0], xy_predicted[:,1], f_predicted, marker='.', label = 'Punti predetti', s=50, color = 'yellow') 
plt.title("Predizione della rete", fontsize='18')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.legend()
plt.show()
```
