e per prima cosa stampo i valori istantanei dell'energia potenziale $U/N$ e della pressione $P$. Nei grafici sottostanti riporto solo i primi 1000 steps per avere una stima di quanti passi mi servono per equilibrare il sistema: l'andamento delle grandezze nelle 3 fasi è diverso ma in tutti i casi dopo circa 200-300 steps la simulazione è equilibrata. Imposto quindi nel codice $n_{eq}= 400$ passi di equilibrazione.

### Autocorrelazione
Utilizzo quindi gli M dati istantanei prodotti per calcolare le funzioni di autocorrelazione di $U/N$ e di $P$. La funzione di autocorrelazione ci da una stima di quanto tempo (in passi MC) impiegano le configurazioni campionate con Metropolis a scorrelarsi una dall'altra e in questo modo ci da un'idea di quanto devono essere grossi i blocchi. La funzione di autocorrelazione è definita come:
$$ AC[\tau] = \chi(\tau)=\frac{\langle \ x(\ t \ ) \cdot x(\ t + \tau \ ) \ \rangle -\langle x( \ t\ ) \rangle \cdot \langle x( \ t + \tau \ ) \rangle }{\sigma_ x^2}$$



```python
epot_s = np.loadtxt("es7.1/MonteCarlo_NVT/solid/autoc_epot.dat",usecols=(0), unpack = True) 
pres_s = np.loadtxt("es7.1/MonteCarlo_NVT/solid/autoc_pres.dat",usecols=(0), unpack = True) 
epot_l = np.loadtxt("es7.1/MonteCarlo_NVT/liquid/autoc_epot.dat",usecols=(0), unpack = True) 
pres_l = np.loadtxt("es7.1/MonteCarlo_NVT/liquid/autoc_pres.dat",usecols=(0), unpack = True) 
epot_g = np.loadtxt("es7.1/MonteCarlo_NVT/gas/autoc_epot.dat",usecols=(0), unpack = True) 
pres_g = np.loadtxt("es7.1/MonteCarlo_NVT/gas/autoc_pres.dat",usecols=(0), unpack = True) 

fig = plt.subplots(1, 2, figsize=(16, 5))
plt.suptitle('Autocorrelazione, solido',fontsize=16)

#Energia potenziale
ax1=plt.subplot(1,2,1)
ax1.plot(epot_s, color="blue")
ax1.set_xlabel('tau (steps)',fontsize=13)
ax1.set_ylabel("AC",fontsize=14);
ax1.set_title("AC per En. Potenziale, solido",fontsize=16)
ax1.grid(True)

#Pressione
ax2=plt.subplot(1,2,2)
ax2.plot(pres_s, color="green")
ax2.set_xlabel('tau (steps)',fontsize=13)
ax2.set_ylabel("AC",fontsize=14);
ax2.set_title("AC per pressione, solido",fontsize=16)
ax2.grid(True)

plt.show()

fig = plt.subplots(1, 2, figsize=(16, 5))
plt.suptitle('Autocorrelazione, liquido',fontsize=16)

#Energia potenziale
ax1=plt.subplot(1,2,1)
ax1.plot(epot_l, color="dodgerblue")
ax1.set_xlabel('tau (steps)',fontsize=13)
ax1.set_ylabel("AC",fontsize=14);
ax1.set_title("AC per En. Potenziale, liquido",fontsize=16)
ax1.grid(True)

#Pressione
ax2=plt.subplot(1,2,2)
ax2.plot(pres_l, color="limegreen")
ax2.set_xlabel('tau (steps)',fontsize=13)
ax2.set_ylabel("AC",fontsize=14);
ax2.set_title("AC per pressione, liquido",fontsize=16)
ax2.grid(True)

plt.show()

fig = plt.subplots(1, 2, figsize=(16, 5))
plt.suptitle('Autocorrelazione, gas',fontsize=16)

#Energia potenziale
ax1=plt.subplot(1,2,1)
ax1.plot(epot_g-0.01, color="deepskyblue")
ax1.set_xlabel('tau (steps)',fontsize=13)
ax1.set_ylabel("AC",fontsize=14);
ax1.set_title("AC per En. Potenziale, gas",fontsize=16)
ax1.grid(True)

#Pressione
ax2=plt.subplot(1,2,2)
ax2.plot(pres_g, color="greenyellow")
ax2.set_xlabel('tau (steps)',fontsize=13)
ax2.set_ylabel("AC",fontsize=14);
ax2.set_title("AC per pressione, gas",fontsize=16)
ax2.grid(True)

plt.show()
```

Come si può vedere dai grafici, nelle 3 diverse fasi si ha un diverso valore di steps necessari per raggiungere AC $\sim 0$:

- nella fase solida l'autocorrelazione decade in circa 100 steps MC e quindi i blocchi dovranno essere almeno qualche volta più grandi di 100;
- nella fase liquida decade in circa 200 steps MC;
- nella fase gassosa invece decade più velocemente, in circa 50 steps MC.

### Incertezza statistica in funzione di L

In questo modo riesco a fare un'analisi più precisa della larghezza necessaria dei blocchi perchè l'andamento che si osserva è che se la taglia del blocco L è troppo piccola l'incertezza statistica è sottostimata; aumentando L l'incertezza cresce fino a quando inizia a saturare a un certo valore limite. Il valore di L a cui arriva a saturazione indica un blocco sufficientemente grande per non sottostimare l'errore. Nei grafici sottostanti, riporto prima l'incertezza statistica in funzione di L in tutto il range $L \in [10,5000]$ e poi in un range minore per determinare meglio a che L si raggiunge il valore limite di saturazione.

## Esercizio 07.4
In questo esercizio utilizzo il codice MC_NVT con le modifiche degli esercizi 07.1 e 07.2 per simulare le tre differenti fasi dell'Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) nelle condizioni:

- fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (raggio di cut-off: $r_c = 2.2$)
- fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ (raggio di cut-off: $r_c = 2.5$)
- fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ (raggio di cut-off: $r_c = 5.0$)

Per ogni fase equilibro il sistema e in seguito con una simulazione di $M=500000$ steps divisi in $N=100$ blocchi (usando quindi $L=5000$ che è adatto per non sottostimare l'incertezza statistica) misuro le quantità $U/N$ e $P$ e calcolo le loro medie e incertezze con il metodo a blocchi, riportando i grafici dalle unità ridotte di L-J alle unità del SI: $$ U= \epsilon U^* \qquad P = \frac{P^* \epsilon}{\sigma^3} $$

In seguito riporto anche nello stesso grafico la funzione di distribuzione radiale $g(r)$ ottenuta dal codice MC e dal codice di dinamica molecolare modificato nell'esercizio 07.3 per confrontarle meglio.

## Solido

```python
kB = 1.380649 * 10**-23 #[J/K]
epsilon_kB = 120    
epsilon = epsilon_kB * kB
sigma=0.34 * 10**-9

x,epot_ave,err_epot = np.loadtxt("es7.4/solid/output.epot.0",usecols=(0,2,3), unpack = True)
x,pres_ave,err_pres = np.loadtxt("es7.4/solid/output.pres.0",usecols=(0,2,3), unpack = True)  

fig = plt.subplots(1, 2, figsize=(15, 5))
plt.suptitle('Simulazione solido',fontsize=16)

#En.Potenziale
ax1=plt.subplot(1,2,1)
ax1.errorbar(x,epot_ave*epsilon,yerr=err_epot*epsilon, color="blue")
ax1.set_xlabel('N blocchi',fontsize=14)
ax1.set_ylabel("En. Pot. [J]",fontsize=14);
ax1.set_title("Energia potenziale media",fontsize=16)
ax1.axhline(y=-7.54659*epsilon, color='black', linestyle='-')
ax1.grid(True)

#Pressione
ax2=plt.subplot(1,2,2)
ax2.errorbar(x,pres_ave*epsilon/(sigma**3),yerr=err_pres*epsilon/(sigma**3), color="green")
ax2.set_xlabel('N blocchi',fontsize=14)
ax2.set_ylabel("P [Pa]",fontsize=14);
ax2.set_title("Pressione media",fontsize=16)
ax2.axhline(y=7.59355*epsilon/(sigma**3), color='black', linestyle='-')
ax2.grid(True)
```
