## Esercizio 07.1

Use the Monte Carlo NVT code we have provided to <span style="color:red">print out instantaneous values of potential energy per particle, $U/N$, and pressure, $P$, along **equilibrated** Monte Carlo simulations</span> (of $M=5\times 10^5$ MC steps and where the Metropolis acceptance ratio has been fixed approximately to 50%) in the following thermodynamic conditions:
- solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
- liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
- gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

1. <span style="color:red">Once this data has been produced, compute and make pictures with a python script of the autocorrelation functions of $U/N$ and $P$ to approximately estimate how large the blocks in your simulations should be.
2. Use the same data to study, with a python script, the estimation of the statistical uncertainties of the expectation value of $U/N$ and $P$ for different size of the blocks in the data blocking technique, from $L=M/N=10$ to $L=5\times 10^3$, being $N$ the number of blocks. <span style="color:red">Make pictures of these statistical uncertainties as a function of $L$.</span> What you observe in the statistical uncertainties after that the block size is large enough (from the point of view of the data blocking technique)?

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

epot_s = np.loadtxt("es7.1/MonteCarlo_NVT/solid/equil/output_epot.dat",usecols=(0), unpack = True) 
pres_s = np.loadtxt("es7.1/MonteCarlo_NVT/solid/equil/output_pres.dat",usecols=(0), unpack = True) 
epot_l = np.loadtxt("es7.1/MonteCarlo_NVT/liquid/equil/output_epot.dat",usecols=(0), unpack = True) 
pres_l = np.loadtxt("es7.1/MonteCarlo_NVT/liquid/equil/output_pres.dat",usecols=(0), unpack = True) 
epot_g = np.loadtxt("es7.1/MonteCarlo_NVT/gas/equil/output_epot.dat",usecols=(0), unpack = True) 
pres_g = np.loadtxt("es7.1/MonteCarlo_NVT/gas/equil/output_pres.dat",usecols=(0), unpack = True) 

fig = plt.subplots(2, 1, figsize=(15, 12))
plt.suptitle('Equilibrazione, valori istantanei',fontsize=16)

#Energia potenziale
ax1=plt.subplot(2,1,1)
ax1.plot(epot_s, color="blue", label = 'solido')
ax1.plot(epot_l, color="dodgerblue", label = 'liquido')
ax1.plot(epot_g, color="deepskyblue", label = 'gas')
ax1.set_xlabel('N misure',fontsize=13)
ax1.set_ylabel("En. Potenziale",fontsize=14);
ax1.set_title("En. Potenziale",fontsize=16)
ax1.legend()
ax1.set_xlim(0,1000)
ax1.grid(True)

#Pressione
ax2=plt.subplot(2,1,2)
ax2.plot(pres_s, color="green",label = 'solido')
ax2.plot(pres_l, color="limegreen",label = 'liquido')
ax2.plot(pres_g, color="greenyellow",label = 'gas')
ax2.set_xlabel('N misure',fontsize=13)
ax2.set_ylabel("Pressione",fontsize=14);
ax2.set_title("Pressione",fontsize=16)
ax2.legend()
ax2.set_xlim(0,1000)
ax2.grid(True)

plt.show()
```

### Exercise 07.4

By means of your upgraded MC code, equilibrate and <span style="color:red">perform MC NVT simulations via a Lennard-Jones model</span> of Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) in the following conditions:
1. solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

show in pictures the obtained average values and uncertainties for the potential energy per particle, $U/N$, the pressure $P$ and the radial distribution function $g(r)$ in SI units ... <span style="color:red"> and compare your MC results for the radial distribution function, $g(r)$, with those obtained with Molecular Dynamics NVE simulations in similar thermodynamic conditions.</span>

## Solido

```python
kB = 1.380649 * 10**-23 #[J/K]
epsilon_kB = 120    
epsilon = epsilon_kB * kB
sigma=0.34 * 10**-9

x,epot_ave,err_epot = np.loadtxt("es7.4/solid/output.epot.0",usecols=(0,2,3), unpack = True)
x,pres_ave,err_pres = np.loadtxt("es7.4/solid/output.pres.0",usecols=(0,2,3), unpack = True)  

fig = plt.subplots(1, 2, figsize=(15, 5))
plt.suptitle('Simulazione solido',fontsize=16)

#En.Potenziale
ax1=plt.subplot(1,2,1)
ax1.errorbar(x,epot_ave*epsilon,yerr=err_epot*epsilon, color="blue")
ax1.set_xlabel('N blocchi',fontsize=14)
ax1.set_ylabel("En. Pot. [J]",fontsize=14);
ax1.set_title("Energia potenziale media",fontsize=16)
ax1.axhline(y=-7.54659*epsilon, color='black', linestyle='-')
ax1.grid(True)

#Pressione
ax2=plt.subplot(1,2,2)
ax2.errorbar(x,pres_ave*epsilon/(sigma**3),yerr=err_pres*epsilon/(sigma**3), color="green")
ax2.set_xlabel('N blocchi',fontsize=14)
ax2.set_ylabel("P [Pa]",fontsize=14);
ax2.set_title("Pressione media",fontsize=16)
ax2.axhline(y=7.59355*epsilon/(sigma**3), color='black', linestyle='-')
ax2.grid(True)
```

```python
r_NVE,gr_ave_NVE,err_NVE = np.loadtxt("es7.3/solid/MolDyn/MolecularDynamics_NVE/output.gave.0",usecols=(1,3,4), unpack = True)
r_NVT,gr_ave_NVT,err_NVT = np.loadtxt("es7.4/solid/output.gave.0",usecols=(1,3,4), unpack = True)  

fig2=plt.figure(figsize=(14, 8))
plt.errorbar(r_NVE,gr_ave_NVE,yerr=err_NVE, color="red", label = 'NVE MolDyn')
plt.errorbar(r_NVT,gr_ave_NVT,yerr=err_NVT, color="gold", label = 'NVT MC')
plt.title("Funzione di distribuzione radiale g(r), solido", fontsize=16)
plt.xlabel('r')
plt.ylabel('g(r)')
plt.grid(True)
plt.legend(fontsize=10)
plt.show()
```
