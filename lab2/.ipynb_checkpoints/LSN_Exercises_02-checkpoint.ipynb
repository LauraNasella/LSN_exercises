## Esercizio 02.1
Attraverso dei **metodi Monte Carlo** è possibile calcolare integrali anche in più dimensioni. L'obiettivo di questo esercizio è calcolare con metodi MC l'integrale 1D della funzione $g(x)=\frac{\pi}{2}\cos(\pi x/2)$:
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$
Questo integrale viene calcolato in due modi:
1. usando un **campionamento uniforme** in $[0,1]$, cioè estraendo uniformemete M punti in $[0,1]$; in essi viene poi valutata la funzione g(x) per arrivare alla stima dell'integrale come I= $\frac{1}{M} \sum_{i=1}^M g(x_i)$.

2. usando l' **importance sampling**, cioè campionando sempre in $[0,1]$ una distribuzione di probabilità diversa da quella uniforme, più simile alla funzione integranda g(x). La funzione d(x) deve avere delle caratteristiche precise, come $d(x) \geq 0$ in $[0,1]$ e deve essere normalizzata a 1. Può essere scelta in diversi modi, ma il più semplice è forse quello di una funzione lineare y = ax+b. Imponendo che la retta passi per i punti (0;$\pi/2$) e (1,0) si trova: 
$$ a = -\pi/2, b = \pi/2 \rightarrow y = -\pi/2 x+\pi/2$$
Normalizzandone l'integrale tra [0,1] si trova infine:
$$d(x) = -2x+2$$
Utilizzando il *metodo della trasformata inversa* descritto nella scorsa esperienza di laboratorio, si trova che 
$x=1-\sqrt{1-r}$ con r variabile casuale distribuita uniformemente. La stima dell'integrale è quindi ottenuta come I= $\frac{1}{M} \sum_{i=1}^M \frac{g(x_i)}{d(x_i)}$.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10,7))
x = np.arange(0.0, 1.0, 0.01)
f1 = np.cos(np.pi*x/2)*np.pi/2
f2 = -2*x+2
plt.plot(x,f1, color = 'r', label = "g(x), integranda")
plt.plot(x,f2, color = 'g', label = "d(x), Importance Sampling")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(y=1, color='y', linestyle='-', label = "Campionamento uniforme")
plt.legend()
plt.title("Confronto")
plt.show()
```


In entrambi i casi sono utilizzati $M = 10^4$ lanci e le incertezze statistiche sono calcolate con il metodo a blocchi, con $N=100$ blocchi. Come si può osservare dal grafico sottostante, l'incertezza statistica nell'importance sampling diminuisce notevolmente.

```python
x,y,err = np.loadtxt("es1/I_uniforme.out",usecols=(0,1,2), unpack = True) 
x1,y1,err1 = np.loadtxt("es1/I_importance.out",usecols=(0,1,2), unpack = True) 
fig = plt.figure(figsize=(12,8))

plt.errorbar(x,y,yerr=err, color = 'yellow', label = "Uniform Sampling");
plt.errorbar(x1,y1,yerr=err1, color = 'green', label = "Importance Sampling");
plt.xlabel('N blocchi',fontsize='14')
plt.ylabel('I',fontsize='14');
plt.grid(True)
plt.axhline(y=1, color='r', linestyle='-', label = "Integral")
plt.title("Stima MC di I", fontsize='16')
plt.legend()
plt.show()
```

## Esercizio 02.2 - Random Walk 3D

Questo esercizio consiste nel simulare un **Random Walk** (RW) 3D sia su un reticolo cubico che nel continuo. Il RW consiste in una traiettoria di step successivi casuali, scelti secondo una certa distribuzione di probabilità: esso è comunque caratterizzabile come **processo di Markov** in quanto il passo n-esimo del RW dipende soltanto dal passo (n-1)-esimo. Facendo una trattazione formale (ad esempio in 1D, con passo di lunghezza a) nel limite continuo  con $a \rightarrow 0$ e step temporale $\tau \rightarrow 0$, si trova che la probabilità di trovarsi nel punto x al tempo t (con x e t finiti) è descritta dall'equazione di diffusione:

$$ \frac{\partial p_1(x,t)}{\partial t} = D \frac{\partial^2 p_1(x,t)}{\partial x^2}$$

con $D = \frac{a^2}{2\tau}$. Risolvendo questa equazione differenziale alle derivate parziali si trova che:
$$p_1(x,t)= \frac{1}{\sqrt{4\pi D t}} e^{\frac{-x^2}{4Dt}}$$ 
ovvero una gaussiana con $\mu=0$ e $\sigma^2=2Dt$.

Se considero quindi un RW 1D di N passi, la posizione al passo N-esimo sarà data da $S_N = \sum x_i$, con $x_i$ variabile casuale con una certa densità di probabilità: per il **teorema del limite centrale**, per $N \rightarrow \infty$, $S_N$ tenderà ad una distribuzione gaussiana con media $<S_N>=N \mu = 0$ e varianza $\sigma^2_{S_N} = N \sigma^2$. L'andamento della varianza e quindi della deviazione standard $\sigma = k \sqrt{N} = a \sqrt{N}$, indica il comportamento diffusivo del RW.

### 1. RW su reticolo cubico
In questo caso viene simulato per $10^4$ volte un RW 3D su un reticolo cubico con passo a = 1: ogni RW è costituito da $n_{steps} = 10^2$. Ad ogni passo viene scelta casualmente la direzione tra x,y,z e, scelta la direzione, viene scelto casualmente anche il verso.

I $10^4$ RWs sono poi divisi in $N_{blocchi} = 10^2$ blocchi per stimare le incertezze statistiche con il metodo a blocchi. Dato che nel grafico sottostante è riportata la quantità $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ in funzione nel numero di step i $\in [0,10^2]$, per trovare la sua incertezza è stato necessaro propagarla da quella di $\langle |\vec{r}_N|^2 \rangle_{RW}$.

```python
#RW su un reticolo cubico 3D di a=1 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

def f(N,k,p):  # defining the fitting function, p è l'indice della radice
    return k * (N)**p  

x,y,err = np.loadtxt("es2/parte1/vettore_rN.out",usecols=(0,1,2), unpack = True) 
fig = plt.figure(figsize=(12,8))

N = np.linspace(0,99,100,endpoint=True)
p_opt, p_cov = curve_fit(f, N, y)
y_fit = f(N,p_opt[0],p_opt[1])

plt.errorbar(x,y,yerr=err, color = 'blue', label = "Reticolo cubico", linewidth=1.0);
plt.plot(x,y_fit, color = "red",label = "Fit",linewidth=2.0) # plotting fitted function
plt.xlabel('step i', fontsize=14)
plt.ylabel('$\sqrt{<|r|^2>_{RW}}$',fontsize=14);
plt.grid(True)
plt.title('$\sqrt{<|r|^2>_{RW}}$', fontsize=16)
plt.legend()
plt.show()

print("Optimized parameters [k,p] =")
print(p_opt)
print("Parameters uncertainty =")
print(np.sqrt(np.diagonal(p_cov)))
```

I parametri ottenuti dal fit sono $k = 1.0091305 \sim 1 = a$ e $ p = 0.49840233 \sim 0.5$ ed essi quindi confermano il comportamento diffusivo.

### 1. RW nel continuo
In questo caso viene simulato per $10^4$ volte un RW 3D nel continuo con passo a = 1: ogni RW è costituito da $n_{steps} = 10^2$. Ad ogni passo viene scelta casualmente una direzione campionando uniformemente l'angolo solido:  $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$.

I $10^4$ RWs sono poi divisi in $N_{blocchi} = 10^2$ blocchi per stimare le incertezze statistiche con il metodo a blocchi. Dato che nel grafico sottostante è riportata la quantità $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ in funzione nel numero di step i $\in [0,10^2]$, per trovare la sua incertezza è stato necessaro propagarla da quella di $\langle |\vec{r}_N|^2 \rangle_{RW}$.

```python
#RW nel continuo
x,y,err = np.loadtxt("es2/parte2/vettore_rN.out",usecols=(0,1,2), unpack = True) 
fig = plt.figure(figsize=(12,8))

N = np.linspace(0,99,100,endpoint=True)
p_opt, p_cov = curve_fit(f, N, y)
y_fit = f(N,p_opt[0],p_opt[1])

plt.errorbar(x,y,yerr=err, color = 'blue', label = "Continuo",linewidth=1.0);
plt.plot(x,y_fit, color = "red", label = "Fit",linewidth=2.0) # plotting fitted function
plt.xlabel('step i',fontsize=14)
plt.ylabel('$\sqrt{<|r|^2>_{RW}}$',fontsize=14);
plt.grid(True)
plt.title('$\sqrt{<|r|^2>_{RW}}$',fontsize=16)
plt.legend()
plt.show()

print("Optimized parameters [k,p] =")
print(p_opt)
print("Parameters uncertainty =")
print(np.sqrt(np.diagonal(p_cov)))
```

I parametri ottenuti dal fit sono $k = 1.00621663 \sim 1 = a$ e $ p = 0.49815439 \sim 0.5$ ed essi quindi confermano il comportamento diffusivo.
