In Statistical Mechanics the one-dimensional (1D) **Ising Model** is exactly resoluble. The Hamiltonian for a system of N spins with nearest neighbors interaction is given by (in the following I will assume units such that $\mu_B=1$ and $k_B=1$):

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

where $\{ s_i \}_{i=1,N}$ are spin variables taking values in $\{-1;1\}$, $J>0$ (ferromagnetic model),and with the assumption $s_{N+1} = s_{1}$.

The partition function for $N$ spins, $Z$, can be exactly computed and it turns out to be:

$$
Z=\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \lambda_1^N + \lambda_2^N
$$

where
$\lambda_{1,2} = 
e^{\beta J} 
\cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$, 
and where $\beta=\frac{1}{k_b T} = \frac{1}{T}$.
Once the partition function has been computed, the Helmholtz free energy is readly obtained as $A = -k_B T \ln Z$. All the thermodynamic quantities of interest can be deduced from $A$ or, equivalently, from $Z$:

- the **Internal Energy** per spin (with $h=0$):

$$
\frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}
= \frac{1}{N} \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}
= -\frac{1}{N}\frac{\partial \ln Z}{\partial \beta}
\underset{h=0}= -J \tanh (\beta J) \Biggl( \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
plt.plot(T, e)
plt.title('Ising 1D, internal energy')
plt.xlabel('T')
plt.ylabel('U/N')
plt.show()
```

- the **Heat Capacity** (with $h=0$):

$$
C(N,T) = \frac{\partial U(N,T)}{\partial T}
= \frac{\partial \beta}{\partial T}\frac{\partial U}{\partial \beta}
= -k_B\beta^2 \frac{\partial}{\partial \beta} \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z} \right] = \\
= -k_B\beta^2 \left[ -\sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \sum_{\{ s_i \}} H \exp \left( -\beta H\right)\frac{\partial Z^{-1}}{\partial \beta} \right] = \\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]\frac{1}{Z}\frac{\partial Z}{\partial \beta} \right\} =\\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} - \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]^2 \right\} =
k_B\beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right)
$$

$$
\frac{C(N,T)}{N} = \frac{1}{N}\frac{\partial U(N,T)}{\partial T} =\\
\underset{h=0}= 
k_B (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)}
- N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}
$$

```python
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
plt.plot(T, heat)
plt.title('Ising 1D, Heat capacity')
plt.xlabel('T')
plt.ylabel('C')
plt.show()
```

- The **magnetization** (with $h \not= 0$):

Note that with $h=0$ the magnetization $M(N,T,h=0)=\langle \sum_{i=1,N}s_i \rangle =0$ because in 1D we cannot have a (ferromagnetic) phase transition in absence of an external magnetic field.

$$
M(N,T,h) = -\frac{\partial A}{\partial h} = k_B T \frac{\partial \ln Z}{\partial h} =
\frac{1}{\beta} \frac{1}{Z}\frac{\partial}{\partial h}\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} = \left\langle \sum_{i=1,N}s_i \right\rangle
$$

The exact formula from $M(N,T,h) = k_B T \dfrac{\partial \ln Z}{\partial h}$ is:
$$
\frac{M(N,T,h)}{N}=\frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]
$$

```python
h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.plot(T, M)
plt.title('Ising 1D, magnetization M with h = 0.02')
plt.xlabel('T')
plt.ylabel('$M$')
plt.show()
```

- The **magnetic susceptibility** (with $h=0$):

$$
\chi(N,T) = \frac{\partial M}{\partial h} = 
\frac{\partial}{\partial h}\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} =\\
= \beta\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right]^2 \frac{\exp \left( -\beta H\right)}{Z}
-\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z}
\frac{1}{Z}\frac{\partial Z}{\partial h} = 
\beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle - \left\langle \sum_{i=1,N}s_i \right\rangle^2 \right] = \\
\underset{h=0}= \beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle \right] = \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

```python
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
plt.plot(T, X)
plt.title('Ising 1D, Susceptibility')
plt.xlabel('T')
plt.ylabel('$\chi$')
plt.show()
```

## Esercizio 06.1
Il codice fornitoci performa una simulazione del modello di Ising 1D con $N=50$ spin. In questo esercizio completiamo il codice con gli algoritmi di Metropolis e di Gibbs per campionare il peso di Boltzmann $p(\{ s_i \}) = \exp(-\beta H)/Z$, aggiungendo anche il calcolo di altre grandezze oltre all'energia interna $U(N,T)$ come la capacit termica $C(N,T)$, la suscettivit magnetica $\chi(N,T)$ e la magnetizzazione $M(N,T,h)$ e introducendo la possibilit di ripartire da una configurazione di spin precedente.

### Algoritmo di Metropolis
Il primo algoritmo che usiamo per campionare il peso di Boltzmann $p(\{ s_i \}) = \exp(-\beta H)/Z$  quello di Metropolis. L'algoritmo  composto dai seguenti step che devono essere ripetuti un grande numero di volte M, m = 1,..,M:

- Viene scelto casualmente uno dei 50 spin $s_i$ dalla configurazione iniziale $\mu$: da esso propongo la mossa di "flipparlo", ovvero se $s_i = -1$ dopo la mossa esso  $s_i = +1$ e viceversa, mentre tutti gli altri spin restano uguali; questa nuova configurazione con lo spin i-esimo "flippato"  chiamata $\nu$;
- Devo quindi decidere se accettare la mossa proposta. L'accettazione  data da $\alpha = A(\nu|\mu)=\mathrm{min} [1,\frac{p(\nu)}{p(\mu)}]= \mathrm{min} [1,e^{-\beta (E_{\nu}-E_{\mu})}]$. Valutando quindi la differenza di energia tra le due configurazioni, genero un numero casuale r in $[0,1]$ e se $r \leq \alpha$ ruoto lo spin, cio accetto la mossa, altrimenti tengo la vecchia configurazione $\mu$.

In questo caso quindi **non** siamo noi a poter regolare l'accettazione con la regola empirica del 50% e quindi lavorare a temperature molto basse diventa complicato per l'accettazione molto piccola.

### Algoritmo di Gibbs
Possiamo vedere l'algoritmo di Gibbs come un Metropolis in cui l'accettazione  sempre $\alpha=1$, cio ogni mossa proposta viene accettata. Nel caso del modello di Ising 1D viene scelto casualmente uno dei 50 spin $s_k$ e si calcola la differenza di energia tra le 2 configurazioni in cui quello spin  +1 o -1, indipendentemente dalla configurazione in cui effettivamente era il sistema:
$$ p(s_k=\pm1)=\frac{1}{1+ \text{exp}\left(\mp \ 2 \beta [J(s_{k-1} + s_{k+1})]\right)}$$
In questo modo viene quindi scelta la nuova configurazione indipendentemente dalla vecchia.

### Ripartenza
Dopo aver aggiunto i due algoritmi ho modificato il codice per permettere la ripartenza da una configurazione di spin precedente:

- Per prima cosa nel file input.dat  bene scegliere il campo h e l'algoritmo da utilizzare durante la simulazione: se voglio utilizzare il Metropolis devo porre la variabile metro = 1, mentre altrimenti si usa Gibbs. Ho quindi aggiunto al file input.dat un'altra variabile restart che pu assumere valori 0 o 1;
- Prima di poter far ripartire il codice da una configurazione precedente, esso deve girare almeno una volta come ci  stato fornito, cio preparando gli spin iniziali in modo casuale. Per poter far girare il codice in questa fase iniziale, controllo che la variabile restart sia posta = 0 ed elimino i files ottenuti in simulazioni precedenti con ./clean.sh. Eseguo quindi il programma ./Monte_Carlo_ISING_1D.exe. 
- Alla fine di questa prima esecuzione la configurazione di spin finale viene salvata nel file config.final; se adesso modifico il file input.dat ponendo la variabile restart = 1, il codice ripartir da quest'ultima configurazione. 

### Equilibrazione
Sfrutto quindi la ripartenza per separare la fase di equilibrazione da quella di effettiva simulazione. Per prima cosa faccio girare il codice con restart=0 a $T = 2.0$: in questo modo il sistema si equilibra a questa temperatura in poche decine di step, dato che ogni step MC consiste nel provare a "flippare" una volta tutti gli spins del sistema. Ci  visibile dal grafico a sinistra in cui  riportato il valore istantaneo dell'energia interna nei primi 500 steps MC e si pu gi vedere come essa fluttua intorno al valore predetto dalla teoria. L'effettiva fase di simulazione con le stime delle grandezze $U(N,T)$, $C(N,T)$, $\chi(N,T)$, $M(N,T,h)$ si ha quando faccio ripartire il codice (restart=1). Per ognuna delle 30 temperature $T$ in $[0.5;2.0]$ considero M = 400000 steps MC divisi in 40 blocchi (ognuno quindi da $10^4$ steps): parto da $T=2$ e gradualmente la abbasso fino a $T=0.5$, utilizzando come configurazione iniziale di una certa temperatura la configurazione di spins finale della temperatura precedente (pi alta). In questo modo il numero di step per equilibrare il sistema  minore rispetto a quello che servirebbe se facessi partire ad ogni nuova T il sistema da una configurazione di spin casuale. Nel grafico di sotto a destra si pu infatti osservare come in questo modo, cio partendo dalla configurazione precedente, per la magnetizzazione a $T=0.5$ (che  la pi bassa a cui simuliamo il sistema) siano sufficienti circa 100 steps per equilibrare il sistema.

```python
y = np.loadtxt("Ising_1D/ISING_1D/equil.ene.metro",usecols=(2), unpack = True) 
m = np.loadtxt("Ising_1D/ISING_1D/equil.mag.metro",usecols=(2), unpack = True) 
y1 = np.loadtxt("Ising_1D/ISING_1D/equil.ene.gibbs",usecols=(2), unpack = True)  

fig = plt.subplots(1, 2, figsize=(15, 6))
plt.suptitle('Equilibrazione',fontsize=16)

nspin = 50

#Energia
ax1=plt.subplot(1,2,1)
ax1.plot(y/nspin, color="blue", label="Metropolis")
ax1.plot(y1/nspin, color='deepskyblue', label="Gibbs")
ax1.set_xlabel('N steps',fontsize=14)
ax1.set_ylabel("U(N,T)/N",fontsize=14);
ax1.set_title("Equilibrazione energia interna U(N,T) a T=2",fontsize=16)
ax1.set_xlim(0, 500)
ax1.axhline(y=-0.46, color='yellow', linestyle='-',label="Teoria")
ax1.legend()
ax1.grid(True)

#Magnetizzazione
ax2=plt.subplot(1,2,2)
ax2.set_xlabel('N steps',fontsize=14)
ax2.set_ylabel("M(N,T,h)",fontsize=14);
ax2.set_title("Equilibrazione magnetizzazione M(N,T,h), h=0.02, T=0.5",fontsize=16)
ax2.axhline(y=1, color='red', linestyle='-', label="Teoria")
ax2.plot(m/nspin, color="green", label = "Metropolis")
ax2.set_xlim(0, 500)
ax2.legend()
ax2.grid(True)
```

## Simulazione
Ho quindi simulato il sistema per le 30 temperature nel range $[0.5,2.0]$, considerando sempre 40 blocchi di $10^4$ steps per calcolare le incertezze statistiche con il metodo a blocchi. Nei grafici sottostanti riporto quindi il valor medio e l'incertezza del 40-esimo blocco per ogni temperatura.

### Energia interna per spin (con $h=0$): $\frac{U(N,T)}{N}=\frac{\langle H \rangle_{h=0}}{N}$

```python
import numpy as np
import matplotlib.pyplot as plt

x,y,err = np.loadtxt("Ising_1D/ISING_1D/Metro/output.ene.metro.0",usecols=(1,3,4), unpack = True) 
x1,y1,err1 = np.loadtxt("Ising_1D/ISING_1D/Gibbs/output.ene.gibbs.0",usecols=(1,3,4), unpack = True) 
fig = plt.figure(figsize=(12,8))

points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
plt.plot(T, e, color = 'yellow',label = "Teoria")
plt.errorbar(x1,y1,yerr=err1, marker='o', color = 'green', label = "Gibbs");
plt.errorbar(x,y,yerr=err,marker='.',color = 'red', label = "Metropolis");
plt.title('Ising 1D, Energia interna per spin', fontsize = '16')
plt.xlabel('T',fontsize='14')
plt.ylabel('U(N,T)/N',fontsize='14')
plt.legend()
plt.grid(True)
plt.xlim(0, 2.5)
plt.show()
```

### Capacit termica per spin, calore specifico (con $h=0$): $\frac{C(N,T)}{N}$, con $C(N,T) = k_B \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$ 

```python
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

x,y,err = np.loadtxt("Ising_1D/ISING_1D/Metro/output.heat.metro.0",usecols=(1,3,4), unpack = True) 
x1,y1,err1 = np.loadtxt("Ising_1D/ISING_1D/Gibbs/output.heat.gibbs.0",usecols=(1,3,4), unpack = True) 
fig = plt.figure(figsize=(12,8))

plt.plot(T, heat, color = 'yellow', label = 'Teoria')
plt.errorbar(x1,y1,yerr=err1, marker = '.',color = 'green', label = "Gibbs");
plt.errorbar(x,y,yerr=err, marker = '.',color = 'red', label = "Metropolis");
plt.title('Ising 1D, Calore specifico', fontsize = '16')
plt.xlabel('T',fontsize='14')
plt.ylabel('C(N,T)/N',fontsize='14')
plt.legend()
plt.grid(True)
plt.xlim(0, 2.5)
plt.show()
```

### Suscettivit magnetica per spin (con $h=0$): $\frac{\chi(N,T)}{N}$ , con  $\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0}$

```python
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

x,y,err = np.loadtxt("Ising_1D/ISING_1D/Metro/output.chi.metro.0",usecols=(1,3,4), unpack = True) 
x1,y1,err1 = np.loadtxt("Ising_1D/ISING_1D/Gibbs/output.chi.gibbs.0",usecols=(1,3,4), unpack = True) 
fig = plt.figure(figsize=(12,8))

plt.plot(T, X, color = 'yellow', label = 'Teoria')
plt.errorbar(x1,y1,yerr=err1, marker = 'o',color = 'green', label = "Gibbs");
plt.errorbar(x,y,yerr=err, marker = '.',color = 'red', label = "Metropolis");
plt.title('Ising 1D, Suscettivit magnetica', fontsize = '16')
plt.xlabel('T',fontsize='14')
plt.ylabel('$\chi$(N,T)/N',fontsize='14')
plt.legend()
plt.grid(True)
plt.xlim(0, 2.5)
plt.ylim(0, 100)
plt.show()
```

### Magnetizzazione per spin (con $h=0.02$): $\frac{M(N,T,h)}{N}$, con $M(N,T,h=0.02)= \langle \sum_{i=1,N}s_i \rangle_{h=0.02}$

```python
h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

x,y,err = np.loadtxt("Ising_1D/ISING_1D/Metro/output.mag.metro.0",usecols=(1,3,4), unpack = True) 
x1,y1,err1 = np.loadtxt("Ising_1D/ISING_1D/Gibbs/output.mag.gibbs.0",usecols=(1,3,4), unpack = True) 
fig = plt.figure(figsize=(12,8))

plt.plot(T, M, color = 'yellow', label = 'Teoria')
plt.errorbar(x1,y1,yerr=err1, marker = '.',color = 'green', label = "Gibbs");
plt.errorbar(x,y,yerr=err, marker = '.',color = 'red', label = "Metropolis");
plt.title('Ising 1D, Magnetizzazione con h = 0.02', fontsize = '16')
plt.xlabel('T',fontsize='14')
plt.ylabel('M(N,T,h)/N',fontsize='14')
plt.legend()
plt.grid(True)
plt.xlim(0, 2.5)
plt.show()
```

Come si pu osservare dai grafici di tutte e 4 le grandezze, i risultati seguono bene le curve teoriche e le barre d'errore sono spesso troppo piccole per essere distinte dal marker del grafico, senza grandi differenze tra l'algoritmo di Metropolis e di Gibbs.
