## Esercizio 03.1 <span style="color:red"> Plain vanilla option pricing </span>
In questo esercizio calcoliamo il prezzo dell'opzione Europea sia nel caso Call $C[S(0),0]$ che nel caso Put $P[S(0),0]$ al tempo t=0.
Un'**opzione** è un contratto tra due parti in cui una delle due parti assume un obbligo e l'altra un diritto. Nel caso dell'opzione Europea (anche detta "plain vanilla") ci possono essere due casi:

- **Call**: Il venditore dell'opzione (*writer*) garantisce al compratore dell'opzione (*holder*) il diritto a comprare un bene con un certo prezzo S(t) effettivo ad un prezzo pattuito K al tempo futuro T. Il *writer* ha quindi l'obbligo di vendere quel bene al prezzo deciso K al tempo T;
- **Put**: In questo caso il *writer* garantisce all'*holder* il diritto a vendere un bene con un prezzo attuale S(t) al prezzo pattuito K al tempo T e quindi il *writer* si assume l'obbligo di comprare.

L'holder eserciterà il suo diritto solo se ne ha profitto, ovvero nel caso **call** solo se $S(T)>K$ (altrimenti compra il bene direttamente dal mercato) e nel caso **put** solo se $S(T)<K$. Il profitto è quindi dato da:

- **Call**: $profitto = max[0,S(T)-K]$;
- **Put**: $profitto = max[0,K-S(T)]$;

Bisogna quindi calcolare il prezzo dell'opzione a t=0 che è il valore di aspettazione del profitto portato indietro al tempo del contratto, cioè:

- **Call**: $C[S(0),0]= \langle e^{-rT}max[0,S(T) - K] \rangle$;
- **Put**: $P[S(0),0]= \langle e^{-rT}max[0,K - S(T)] \rangle$.

Per calcolare questi prezzi è quindi necessario calcolare S(T), cioè il valore effettivo del bene al tempo T finale.

### Soluzione analitica con la teoria di Black-Scholes
La teoria di Black-Scholes assume che l'evoluzione temporale del prezzo del bene sia stocastica e segua un moto Browniano geometrico (GBM) con rate di interesse costante r (drift del GBM) e volatilità $\sigma$. La soluzione analitica di Black-Scholes per il prezzo per t<T è:

- prezzo dell'opzione Europea Call:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- prezzo dell'opzione Europea Put:
$$ P\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

dove 
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right]$$

dove $d_2 = d_1 - \sigma\sqrt{T-t}$, e dove
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} = \frac{1}{2} \left[ 1+ {\rm erf} \left( \frac{x}{\sqrt{2}}\right) \right]$$
è la cumulativa della Gaussiana $\sim \mathcal{N}(\mu=0,\sigma^2=1)$,
e dove
$${\rm erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x} dz \, e^{-z^2}$$

Ad esempio usando i parametri
- prezzo a $t=0$: $S(0)=100$
- tempo finale: $T=1$
- prezzo pattuito: $K=100$
- rate di interesse: $r=0.1$
- volatilità: $\sigma=0.25$

si ottiene:

```python
import numpy as np
import scipy
from scipy.special import erf
from numpy import log, exp, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

### Monte Carlo
In questo esercizio, calcoliamo il prezzo dell'opzione Europea sia nel caso call $C[S(0),0]$ sia nel caso put $P[S(0),0]$ a t=0 calcolando S(T) in due diversi modi:

1) Campionando **direttamente** il prezzo finale $S(T)\sim GBM(r,\sigma^2)$ con valore iniziale $S(0)$ come:
    $$S(T) = S(0) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)T+\sigma W(T)\right]$$
    dove $W(t)\sim\mathcal{N}(0,T)$.
    
2) Campionando il $GBM(r,\sigma^2)$ **discrettizandolo**, ovvero dividendo $[0,T]$ in $10^2$ intervalli temporali: $S(0)$, $S(t_1)$, $S(t_2)$, ... fino a $S(t_{100}=T=1)$. Ad ogni tempo successivo, il prezzo dipende solo dall'istante precedente, con incrementi di $W$ indipendenti e distribuiti normalmente. Quindi per i = 1,...,n si ha:
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    con $Z_i \sim \mathcal{N}(0,1)$.
    
Nella simulazione ho usato i parametri
- prezzo a $t=0$: $S(0)=100$
- tempo finale: $T=1$
- prezzo pattuito: $K=100$
- rate di interesse: $r=0.1$
- volatilità: $\sigma=0.25$

e ho considerato $M = 10^5$ simulazioni divise in N = 100 blocchi per stimare le incertezze statistiche con il metodo a blocchi.

### Call

```python
import matplotlib
import matplotlib.pyplot as plt
x,y,err = np.loadtxt("parte1/Call_diretto.out",usecols=(0,1,2), unpack = True) 
x1,y1,err1 = np.loadtxt("parte2/Call_discreto.out",usecols=(0,1,2), unpack = True) 
fig = plt.figure(figsize=(12,8))

plt.errorbar(x,y,yerr=err, color = 'yellow', label = "Campionamento diretto");
plt.errorbar(x1,y1,yerr=err1, color = 'green', label = "Campionamento discreto");
plt.xlabel('N blocchi',fontsize='14')
plt.ylabel('C[S(0),0]',fontsize='14');
plt.grid(True)
plt.axhline(y=14.975790778311286, color='r', linestyle='-', label = "Sol. analitica di Black-Scholes")
plt.title("Stima MC di C[S(0),0]", fontsize='16')
plt.legend()
plt.show()
```

### Put

```python
x,y,err = np.loadtxt("parte1/Put_diretto.out",usecols=(0,1,2), unpack = True) 
x1,y1,err1 = np.loadtxt("parte2/Put_discreto.out",usecols=(0,1,2), unpack = True) 
fig = plt.figure(figsize=(12,8))

plt.errorbar(x,y,yerr=err, color = 'yellow', label = "Campionamento diretto");
plt.errorbar(x1,y1,yerr=err1, color = 'green', label = "Campionamento discreto");
plt.xlabel('N blocchi',fontsize='14')
plt.ylabel('P[S(0),0]',fontsize='14');
plt.grid(True)
plt.axhline(y=5.4595325819072364, color='r', linestyle='-', label = "Sol. analitica di Black-Scholes")
plt.title("Stima MC di P[S(0),0]", fontsize='16')
plt.legend()
plt.show()
```

In entrambi i casi **call** e **put** e per entrambi nei casi di campionamento diretto e discretizzato, il valore medio converge al valore trovato analaticamente dalle soluzioni di Black-Scholes entro le incertezze statistiche.
