## Esercizio 04.1
Il codice di Dinamica Molecolare (MD) performa una simulazione nell'ensemble microcanonico NVE (N, V ed E sono costanti) di N = 108 particelle in un box di volume V interagenti con potenziale di Lennard-Jones usando l'algoritmo di Verlet che è un metodo di integrazione numerica per integrare le equazioni del moto conservando l'energia E. Ho modificato il codice per permettere la ripartenza, leggendo sia l'ultima configurazione che quella precedente seguendo questi passaggi:

- Per prima cosa nel file input.dat ho inserito come valori $T = 0.8$, $\rho = 1.1 $ e raggio di cut-off $r_c=2.2$ per poter simulare una fase solida e ho inserito un'altra variabile *restart* che può assumere valori 0 o 1;
- Prima di poter far ripartire il codice, esso deve girare almeno una volta come ci è stato fornito, cioè preparando le velocità iniziali in modo casuale con CM nullo. Per poter far girare il codice in questa fase iniziale, controllo che la variabile restart sia posta = 0, elimino i files ottenuti in simulazioni precedenti con ./clean.sh e copio la configurazione fcc in quella iniziale col comando: cp config.fcc config.0. Eseguo quindi il programma ./MolDyn_NVE.exe;
- Ho modificato la funzione ConfFinal per permettere la ripartenza: salvo l'ultima configurazione spaziale $\vec{r}(t)$ nel file config.0 e quella ad essa precedente $\vec{r}(t-dt)$ in old.0. Per attivare effettivamente la ripartenza bisogna porre ora la variabile *restart*=1: in questo modo alla prossima esecuzione del programma con il comando ./MolDyn_NVE.exe, il programma ripartirà dalle ultime due configurazioni. Da esse calcolerà la nuova configurazione $\vec{r}(t+dt)$ con un passo dell'algoritmo di Verlet e da $\vec{r}(t + dt)$ e $\vec{r}(t)$ calcolerà $\vec{v}(t+dt/2)$ e da essa la temperatura attuale $T(t+dt/2)$. Confrontando la temperatura attuale con quella target T* determino una fattore di scala $f_s = \sqrt{T^*/T_{attuale}}$ che mi permette di riscalare le velocità per portare il sistema alla temperatura a cui vogliamo effettivamente simulare il sistema. Con le velocità scalate trovo una "nuova" vecchia configurazione $\vec{r}_{new}(t) = \vec{r}(t+dt) - dt \vec{v}_s$; usando questa e $\vec{r}(t)$ faccio partire la simulazione. 
- D'ora in poi, eseguendo il programma, esso ripartirà ogni volta dalle ultime due configurazioni.

### Equilibrazione
Simulo una fase solida con $T = 0.8$, $\rho = 1.1 $ e raggio di cut-off $r_c=2.2$: per equilibrare il sistema ripeto la simulazione con nstep=1000 steps per più volte, misurando le proprietà termodinamiche ogni 10 steps per evitare misure troppo correlate (quindi in ogni simulazione sono effettuate 100 misure). Di seguito riporto l'andamento della temperatura $T$, dell'energia totale (interna) per particella $E/N$, dell'energia potenziale per particella $U/N$, e dell'energia cinetica per particella $K/N$: osservo che già dopo 400-500 misure (quindi dopo 4-5 ripartenze) le proprietà termodinamiche del sistema si stabilizzano.

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

temp = np.loadtxt("es4.1/MolDyn/MolecularDynamics_NVE/output_temp.dat",usecols=(0), unpack = True)
etot = np.loadtxt("es4.1/MolDyn/MolecularDynamics_NVE/output_etot.dat",usecols=(0), unpack = True) 
epot = np.loadtxt("es4.1/MolDyn/MolecularDynamics_NVE/output_epot.dat",usecols=(0), unpack = True) 
ekin = np.loadtxt("es4.1/MolDyn/MolecularDynamics_NVE/output_ekin.dat",usecols=(0), unpack = True) 

fig = plt.subplots(2, 2, figsize=(15, 11))
plt.suptitle('Equilibrazione',fontsize=16)

#Temperatura
ax1=plt.subplot(2,2,1)
ax1.plot(temp, color="blue")
ax1.set_xlabel('N misure',fontsize=14)
ax1.set_ylabel("T",fontsize=14);
ax1.set_title("Temperatura",fontsize=16)
ax1.axhline(y=0.8, color='deepskyblue', linestyle='-')
ax1.grid(True)

#Energia Totale
ax2=plt.subplot(2,2,2)
ax2.plot(etot, color="green")
ax2.set_xlabel('N misure',fontsize=14)
ax2.set_ylabel("En. Totale",fontsize=14);
ax2.set_title("En. Totale",fontsize=16)
ax2.axhline(y=-5.5, color='lawngreen', linestyle='-')
ax2.grid(True)

#Energia Cinetica
ax3=plt.subplot(2,2,3)
ax3.plot(ekin, color="yellow")
ax3.set_xlabel('N misure',fontsize=14)
ax3.set_ylabel("En. Cinetica",fontsize=14);
ax3.set_title("En. Cinetica",fontsize=16)
ax3.axhline(y=1.2, color='gold', linestyle='-')
ax3.grid(True)

#Energia Potenziale
ax4=plt.subplot(2,2,4)
ax4.plot(epot, color="orange")
ax4.set_xlabel('N misure',fontsize=14)
ax4.set_ylabel("En. Potenziale",fontsize=14);
ax4.set_title("En. Potenziale",fontsize=16);
ax4.axhline(y=-6.7, color='orangered', linestyle='-')
ax4.grid(True)

fig2=plt.figure(figsize=(14, 8))
plt.plot(temp, color = "blue",label='Temperatura')
plt.plot(etot, color = "green",label='En. Totale')
plt.plot(ekin, color = "yellow",label='En. Cinetica')
plt.plot(epot, color = "orange",label='En. Potenziale')
plt.axhline(y=0.8, color='deepskyblue', linestyle='-')
plt.axhline(y=-5.5, color='lawngreen', linestyle='-')
plt.axhline(y=1.2, color='gold', linestyle='-')
plt.axhline(y=-6.7, color='orangered', linestyle='-')
plt.title("Equilibrazione", fontsize=16)
plt.xlabel('N misure')
plt.ylabel('Equilibrazione')
plt.grid(True)
plt.legend(fontsize=10)
plt.show()
```

Dai grafici si nota che la $T$ e l'energia cinetica $K/N$ hanno lo stesso andamento e ciò è dovuto al fatto che sono grandezze proporzionali. Invece l'energia cinetica $K/N$ e quella potenziale $U/N$ hanno un andamento con oscillazioni uguali ed opposte in quanto l'algoritmo conserva l'energia totale che è data dalla loro somma $E = K + U$: le oscillazioni dell'energia totale sono molto minori rispetto alle oscillazioni di $K$ e $U$.

In questo esercizio utilizzo il codice di MD con le modifiche degli esercizi 04.1 e 04.2 per simulare tre differenti fasi dell'**Argon** ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) nelle condizioni:
1. fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (raggio di cut-off: $r_c = 2.2$)
2. fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ (raggio di cut-off: $r_c = 2.5$)
3. fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ (raggio di cut-off: $r_c = 5.0$)

Per ogni fase equilibro il sistema come nell'esercizio 04.1; in seguito con una simulazione di nstep=10000 steps misuro le quantità termodinamiche $T$, $E/N$, $U/N$, $K/N$ e calcolo le loro medie e incertezze con il metodo a blocchi con nblocks=50, riportando i grafici dalle unità ridotte di L-J alle unità del SI:
$$  T = \frac{\epsilon}{k_B}T^* \qquad U= \epsilon U^* \qquad E= \epsilon E^* \qquad K= \epsilon K^* \qquad$$

```python
kB = 1.380649 * 10**-23 #[J/K]
epsilon_kB = 120    
epsilon = epsilon_kB * kB
```

## Esercizio opzionale 04.4 - pressione

Ho infine aggiunto anche il calcolo della pressione: dal teorema del viriale, la pressione può essere calcolata come:

$$ P = \rho k_B T + \frac{1}{3V} \left\langle \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 48\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle $$

Ho quindi equilibrato il sistema per tutte e tre le fasi solida, liquida e gassosa e infine ho calcolato valori istantanei, valori medi e incertezze con il metodo a blocchi, riportando i risultati finali in unità di misura del SI.
