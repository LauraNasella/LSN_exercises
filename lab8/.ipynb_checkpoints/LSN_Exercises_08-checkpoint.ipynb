During this exercise you will variationally optimize the ground state of a single quantum particle in a one dimensional (1D) space confined by the following external potential:
$$
V(x) = x^4 - \frac{5}{2}x^2
$$

This 1D quantum model is not analytically solvable and should be attacked by numerical methods. In order to obtain an approximate wave function for the ground state, the Variational Monte Carlo method uses a trial wave function, parametrized by a set of variational parameters. An approximate model for the ground state wave function is the following trial wave function:
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

$\Psi_T^{\sigma,\mu}(x)$ depends on two variational parameters, $\sigma$, which is the width of the two Gaussians under square root, and $\mu$, which represents the distance from the origin of their average values.

### Supplementary material

How can we describe a wave function as a vector? We set up a lattice of discrete points, spaced by a constant distance $dx$, and record the value of the function at each lattice point. In this way the wave function, $\Psi(x)$ become a vector:

$$
\Psi(x) \to \left[ \Psi(x_1), \Psi(x_2), ..., \Psi(x_N) \right] = 
\left[ \Psi_1, \Psi_2, ..., \Psi_N \right]
$$

The time independent Schrodinger equation 

$$
{\hat H} \Psi(x) = \left( -\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2} + V(x) \right) \Psi(x) = E \Psi(x)
$$

can be written in matrix form. But how do we write the second derivative at a particular point? Let's start with the first derivatives:
$$
\frac{\partial}{\partial x} \Psi\left(x=(x_i+x_{i+1})/2\right) \simeq \frac{\Psi_{i+1}-\Psi_{i}}{dx}\\
\frac{\partial}{\partial x} \Psi\left(x=(x_i+x_{i-1})/2\right) \simeq \frac{\Psi_{i}-\Psi_{i-1}}{dx}
$$

Thus, we can discretize the second derivative in the following way:
$$
\frac{\partial^2}{\partial x^2} \Psi(x=x_i) =
\frac{\frac{\partial}{\partial x} \Psi\left(x=(x_i+x_{i+1})/2\right) - \frac{\partial}{\partial x} \Psi\left(x=(x_i+x_{i-1})/2\right)}{dx} =
\frac{\Psi_{i+1}-2\Psi_{i}+\Psi_{i-1}}{dx^2}
$$

We finally obtain:

$$
\begin{bmatrix}
 \frac{\hbar^2}{m\,dx^2}+V(x_1)& -\frac{\hbar^2}{2m\,dx^2} & 0 & \dots  & 0 \\
-\frac{\hbar^2}{2m\,dx^2} & \frac{\hbar^2}{m\,dx^2}+V(x_2) & -\frac{\hbar^2}{2m\,dx^2} & \dots  & 0 \\
 0                        & -\frac{\hbar^2}{2m\,dx^2} & \frac{\hbar^2}{m\,dx^2}+V(x_3) & \dots  & 0 \\
 \vdots                   & \vdots                    & \vdots & \ddots & \vdots \\
 0                    & 0                & 0 & \dots  & \frac{\hbar^2}{m\,dx^2}+V(x_N)
\end{bmatrix}
\begin{bmatrix}
\Psi_{1} \\
\Psi_{2} \\
\vdots \\
\Psi_{N}
\end{bmatrix}
= E
\begin{bmatrix}
\Psi_{1} \\
\Psi_{2} \\
\vdots \\
\Psi_{N}
\end{bmatrix}
$$

To solve the time independent Schrodinger equation we therefore have to find eigenstates and eigenvalues of the matrix just obtained.

```python
import numpy as np
import matplotlib.pyplot as plt

def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])
print("1st excited state energy: ", E[1])
print("2nd excited state energy: ", E[2])

# Plot a few things
plt.figure(figsize=(8,5))
scale = 0.3
plt.plot(x, scale*V, color="Black", label="Potential") # plot the potential
plt.plot(x,(psi[0])**2)
plt.plot(x,(psi[1])**2)
plt.plot(x,(psi[2])**2)
plt.title("Potential & Probabilities")
plt.xlabel("x")
plt.grid(True)
plt.xlim((-3,3))
plt.ylim((-0.6,0.6))
plt.show()
```

## Esercizio 08.1
In questo esericizio utilizzo l'algoritmo di Metropolis per campionare il modulo quadro della funzione d'onda di prova $|\Psi_T(x)|^2$ per una particella quantistica 1D usando una probabilità di transizione $T(x_{new}|x_{old})$ uniforme. Lo step viene cambiato di volta in volta controllando che l'accettazione risulti sempre $\sim 50$%. Campiono in questo modo $M=10^5$ punti e per ogni punto campionato x salvo anche l'opposto -x sfruttando la simmetria centrale del problema. Il campionamento della densità di probabilità è necessario per calcolare il valore di aspettazione dell'Hamiltoniana sulla funzione d'onda di prova:
$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$
Applico quindi l'hamiltoniana alla funzione d'onda, cioè la parte cinetica (data in pratica dal laplaciano della funzione d'onda) e la parte di energia potenziale. Le stime sono quindi state divise in $N = 100$ blocchi da cui ho valutato il valor medio di $\langle {\hat H} \rangle_T$ e la sua incertezza statistica con il *data blocking*. Come esempio riporto i risultati nel caso $\mu = 0$, $\sigma = 1$ per cui ottengo $\langle {\hat H} \rangle_T = -0.25812 \pm 0.00686$.

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
x,y,err = np.loadtxt("es8.1/ave_H.out",usecols=(0,1,2), unpack = True) 

plt.figure(figsize=(8,6))
plt.errorbar(x,y,yerr=err, color = 'deepskyblue');
plt.xlabel('N blocchi',fontsize='14')
plt.ylabel('$<{\hat H}>$',fontsize='14');
plt.grid(True)
plt.axhline(y=-0.25812, color='blue', linestyle='-')
plt.title("Stima di $< {\hat H} >$ per $\mu=0$ e $\sigma=1$", fontsize='16')
```

## Esercizio 08.2
In questo esercizio modifichiamo il codice MC variazionale per trovare i parametri $\mu$ e $\sigma$ che minimizzano il valore di aspettazione $\langle {\hat H} \rangle_T$ per la $\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+ e^{-\frac{(x+\mu)^2}{2\sigma^2}}$:
$$
\langle {\hat H} \rangle_T = \frac{\langle \Psi_T^{\sigma,\mu}| {\hat H} | \Psi_T^{\sigma,\mu} \rangle}
{\langle \Psi_T^{\sigma,\mu} | \Psi_T^{\sigma,\mu} \rangle} \ge E_0 =
\frac{\langle \Psi_0| {\hat H} | \Psi_0 \rangle}
{\langle \Psi_0 | \Psi_0 \rangle}
$$
I parametri variazionali vengono mossi secondo l'algoritmo del *simulated annealing*: parto da una certa coppia $\mu$ e $\sigma$ e da essi calcolo $E_{old}=\langle {\hat H} \rangle_T$. Genero quindi una nuova coppia $\mu_{new}$ e $\sigma_{new}$ dalla vecchia coppia con un nucleo stocastico uniforme con $step_{\mu}=step_{sigma}=0.05$ e ricalcolo $E_{new}=\langle {\hat H} \rangle_T$. Accetto quindi la nuova coppia con Metropolis con probabilità $p = e^{-\beta \Delta E}$ con $\Delta E = E_{new} - E_{old}$, con una certa temperatura fittizia T. Partendo da una certa temperatura fittizia iniziale $T_i$ abbastanza elevata (con un corrispondente $\beta_i$) e diminuendo gradualmente la temperatura, il sistema raffredda e quindi si muove verso l'energia minima, ovvero verso i parametri $\mu$ e $\sigma$ che la minimizzano meglio. Ripeto quindi questi passaggi con 20 diminuzioni graduali della temperatura da $T_i=10$ a $T_f=0.0035$, facendo $n=100$ steps ad ogni temperatura e per ogni step campionando $M=10^5$ punti per stimare $\langle {\hat H} \rangle_T$ (prendendo per ogni x anche l'opposta -x).

Di seguito riporto l'andamento di $\langle {\hat H} \rangle_T$ in funzione nel numero di iterazioni del SA (cioè 20 temperature per 100 passi per ogni temperatura = 2000 iterazioni) partendo da $\mu=1$ e $\sigma=1$ e la stima di $\langle {\hat H} \rangle_T$ in funzione del numero di blocchi (N=100) per la coppia di parametri $\mu$ e $\sigma$ determinati.

```python
H = np.loadtxt("es8.2/mu_sigma_H.out",usecols=(1), unpack = True)
x,y,err = np.loadtxt("es8.2/ave_H.out",usecols=(0,1,2), unpack = True) 

fig = plt.subplots(1, 2, figsize=(15, 6))
 
#Andamento di H in funzione del numero di iterazioni del SA
ax1=plt.subplot(1,2,1)
ax1.plot(H, color="blue")
ax1.set_xlabel('N iterazioni',fontsize=14)
ax1.set_ylabel("$<{\hat H}>$",fontsize=14);
ax1.set_title("$<{\hat H}>$ in funzione del numero di iterazioni",fontsize=16)
ax1.grid(True)

#Stima di H in funzione del numero di blocchi
ax2=plt.subplot(1,2,2)
ax2.errorbar(x,y,yerr=err, color = 'deepskyblue');
ax2.set_xlabel('N blocchi',fontsize='14')
ax2.set_ylabel('$<{\hat H}>$',fontsize='14');
ax2.grid(True)
plt.axhline(y=-0.441722, color='blue', linestyle='-')
ax2.set_title("Stima di $< {\hat H} >$ per $\mu$ e $\sigma$ ottimizzati", fontsize='16')
ax2.set_ylim(-0.45,-0.43)
```

### Campionamento $|\Psi_T(x)|^2$
Di seguito riporto l'istogramma delle configurazioni campionate e lo comparo con la curva analitica di $|\Psi_T(x)|^2$ per $\mu$ e $\sigma$ ottimizzati e con la soluzione numerica ottenuta trasformando l'equazione di Schrodinger in equazione matriciale.

```python
camp = np.loadtxt('es8.2/campionamento.out',unpack = True)

def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

def sol_analitica(x, mu, sigma):
    psi_t = np.exp(-0.5*pow((x-mu)/sigma,2)) + np.exp(-0.5*pow((x+mu)/sigma,2))
    norm = 2.*np.sqrt(np.pi)*sigma*(1 + np.exp(-(mu/sigma)**2)) 
    return psi_t**2/ norm

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
print("Ground state energy da Schrodinger: ", E[0])

psi2= sol_analitica(x,0.794031,0.632193)

# Plot a few things
plt.figure(figsize=(10,7))
scale = 0.3
plt.hist(camp, bins=100, density=True, color = 'lightblue', label='campionamento')
plt.plot(x,(psi[0])**2, color = 'red', label = 'sol. numerica da Schrodinger')
plt.plot(x, psi2, color = "blue", linestyle='-', label='sol. analitica')
plt.title("$|\Psi_T|^2$", fontsize = 16)
plt.xlabel("x")
plt.ylabel("$|\Psi_T|^2$")
plt.grid(True)
plt.legend()
plt.xlim((-3,3))
#plt.ylim((-0.6,0.6))
plt.show()
```

La soluzione analitica ottenuta da $\Psi_T^{\sigma,\mu}(x)$ con i parametri ottimizzati fitta abbastanza bene l'istogramma ottenuto dal campionamento mentre la soluzione numerica non molto a causa della differenza tra la forma funzionale del ground state utilizzato (la somma di due gassiane) e del ground state dall'equazione di Schrodinger.
